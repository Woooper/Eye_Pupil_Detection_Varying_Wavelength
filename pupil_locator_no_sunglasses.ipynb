```python
eye_df = pd.read_csv("real_values_2.csv",delimiter=";")
ml_table = np.zeros(62)
n = 30       # Nokta sayısı
theta = np.linspace(0, 2*np.pi, n, endpoint=False)  # 0'dan 2π'ye kadar açılar
for i in np.arange(0,len(eye_df)):

    h, k = eye_df['Center (px) (x-axis)'].values[i], eye_df['Center (px) (y-axis)'].values[i]       # Çemberin merkezi
    r = float(eye_df['Diameter (px)'].values[i].replace(",", "."))     # Yarıçap
    x = h + r * np.cos(theta)
    y = k + r * np.sin(theta)
    combined = np.column_stack((x, y)).ravel()
    ml_points = np.hstack((np.array(eye_df['Center (px) (x-axis)'].values[i]),np.array(eye_df['Center (px) (y-axis)'].values[i]),combined))
    ml_table = np.vstack((ml_table,ml_points))

```

```python
ml_table_fin = np.concatenate((np.resize(eye_df.FileName.values,(516,1)),ml_table[1:]),axis=1)
```

```python
pd.DataFrame(ml_table_fin).to_csv('center_circle_coor.csv',sep=',',index=False)
```
